```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#create an array x ranging from [0,1] in 100 steps
x = np.linspace(0,1,100)
```

```python
#define a function that returns exp(x)
def returexpo(x):
    return np.exp(x)
```

```python
#Set y = returexpo(x) function
y = returexpo(x)
```

```python
#Plot x and y
plt.plot(x,y)

#Label x-axis as "Time[milliseconds]" and y-axis as "Awesomeness"
plt.xlabel('Time[milliseconds]')
plt.ylabel('Awesomeness')

#Save the figure
plt.savefig('Session_7_Prompt.png', bbox_inches = 'tight', dpi = 600)
```
